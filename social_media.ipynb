```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('socialmedia.csv')

df.columns = df.columns.str.strip()

bins = [0, 18, 30, 50, 100]
labels = ['Teen (0-18)', 'Young Adult (19-30)', 'Adult (31-50)', 'Senior (51+)']

df['Age Category'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

age_counts = df['Age Category'].value_counts()

most_used_age_category = age_counts.idxmax()
most_used_count = age_counts.max()

print(f"Most used age category: {most_used_age_category} ({most_used_count} users)")
most_used_platform = df['Platform'].value_counts().idxmax()
most_used_count = df['Platform'].value_counts().max()

print(f"Most used platform: {most_used_platform} ({most_used_count} users)")

age_category_platform = df.groupby(['Platform', 'Gender', 'Age Category']).size().unstack()

fig, ax = plt.subplots(figsize=(16, 8))

age_category_platform.plot(kind='bar', stacked=True, colormap='coolwarm', edgecolor='black', ax=ax)

plt.title('Platform Usage by Gender and Age Category (Stacked Bar Chart)', fontsize=18, fontweight='bold')
plt.xlabel('Platform', fontsize=14)
plt.ylabel('Number of Users', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title="Age Category", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12)

plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('socialmedia.csv')

device_data = df['DeviceType'].value_counts()
os_data = df['OS'].value_counts()
connection_data = df['ConnectionType'].value_counts()

def plot_pie_chart(data, title):
    labels = data.index
    sizes = data.values
    colors = ['#ff9999', '#66b3ff', '#99ff99', '#c2c2f0', '#ffb3e6']


    explode = [0.1] + [0] * (len(labels) - 1)
    plt.pie(sizes, labels=labels, colors=colors[:len(labels)], autopct='%1.1f%%', startangle=140, explode=explode)
    plt.title(title)

plt.figure(figsize=(18, 6))

plt.subplot(131)
plot_pie_chart(device_data, 'Device Type Distribution')

plt.subplot(132)
plot_pie_chart(os_data, 'Operating System Distribution')

plt.subplot(133)
plot_pie_chart(connection_data, 'Connection Type Distribution')

plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

df = pd.read_csv('socialmedia.csv')

income_bins = [0, 30000, 70000, np.inf]
income_labels = ['Low', 'Medium', 'High']

df['Income Group'] = pd.cut(df['Income'], bins=income_bins, labels=income_labels)

grouped_data = [df[df['Income Group'] == group]['Total Time Spent'] for group in income_labels]

f_stat, p_val = stats.f_oneway(*grouped_data)

print(f"F-statistic: {f_stat}")
print(f"P-value: {p_val}")

if p_val < 0.05:
    print("There is a significant difference in Total Time Spent across different Income Groups.")
else:
    print("There is no significant difference in Total Time Spent across different Income Groups.")
```

```python
import pandas as pd
import scipy.stats as stats

df = pd.read_csv('socialmedia.csv')

correlation, p_value = stats.pearsonr(df['Age'], df['Total Time Spent'])

print(f"Pearson Correlation: {correlation}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("There is a significant negative correlation between Age and Total Time Spent.")
else:
    print("There is no significant correlation between Age and Total Time Spent.")
```
